#286 ** Você está dirigindo um pouco rápido demais, e um policial para você. Escreva uma função para retornar um dos 3 resultados possíveis: "Sem Multa", "Pequena Multa" ou "Grande Multa". Se a sua velocidade for igual ou inferior a 60, o resultado é "Sem Multa". Se a velocidade for entre 61 e 80 inclusive, o resultado é "Pequena Multa". Se a velocidade é de 81 ou mais, o resultado é "Multa Grande". A menos que seja seu aniversário (codificado como um valor booleano nos parâmetros da função) - em seu aniversário, sua velocidade pode ser 5% maior em todos os casos. **

```python
def capturar_velocidade(velocidade, aniversario):
    if aniversario == True:
        velocidade = velocidade + velocidade * 0.05
    if velocidade <= 60:
        return 'Sem Multa'
    elif velocidade <= 80:
        return 'Pequena Multa'
    else:
        return 'Multa Grande'
```

```python
capturar_velocidade(65,False)
```
